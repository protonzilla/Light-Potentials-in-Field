```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
kpc = 5 # relative value for intrinsic rate constant for photochemistry
kdf = 1 # relative value for the sum of intrinsic rate constants for fluorescence and nonradiative decay
kNPQ = 1 # relative value for intrinsic rate constant for NPQ, which will be multiplied by the "concentration"
         # of the NPQ quencher
qL = 1 # The fraction of PSII centers with oxidized QA
NPQ = 0 # The concentration of the NPQ quenchers


def phi2 (qL, NPQ_array): # function that returns an array of phi2 values, given a value for qL and
                          # a range of values for the concentration NPQ.
                          # The function is derived from the Stern-Volmer model for photochemical 
                          # and fluorescence yields, e.g. Kramer et al. (2004)
    phi2_array = [] 
    for NPQ in NPQ_array:
        phi2_array.append(qL*kpc/(kdf + NPQ * kNPQ + qL*kpc ))
    return phi2_array

# phi2 (1, [0,.2,.6])
```

```python

NPQ = np.linspace(0,4,20) # set up a range of 20 NPQ quencher concentrations from 0 to 4

plt.figure() # make a graph

for qL in [0.1, 0.2, 0.4, 0.6, 0.8, 1]:  # calculate six curves, each with a different QA redox state
        plt.plot(NPQ, phi2(qL, NPQ), label = str(qL)) # plot Phi2 versus NPQ

plt.xlabel("NPQ")
plt.ylabel("Phi2")
plt.legend(title = 'Fraction QA oxidized (qL)')
plt.show()

```

```python

NPQ = np.linspace(0,4,20) # set up a range of 20 NPQ quencher concentrations from 0 to 4

plt.figure() # make a graph

for qL in [0.1, 0.2, 0.4, 0.6, 0.8, 1]:  # calculate six curves, each with a different QA redox state
        Phi2_array = phi2(qL, NPQ)
        Phi2_array = Phi2_array/Phi2_array[10]
        plt.plot(NPQ - 2.0, Phi2_array, label = str(qL)) # plot Phi2 versus NPQ

plt.xlabel("Change in NPQ from 2.0")
plt.ylabel("Frational change in Phi2")
plt.legend(title = 'Fraction QA oxidized (qL)')
plt.show()
```
